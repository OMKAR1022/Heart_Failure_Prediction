```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

# so we are applying z-score

## for 3 column RestingBP,Cholesterol,MaxHR

```python
df[df.RestingBP>(df.RestingBP.mean()+3*df.RestingBP.std())] 
```

```python
df[df.RestingBP<(df.RestingBP.mean()-3*df.RestingBP.std())] 
```

```python
df1 = df[df.RestingBP<=(df.RestingBP.mean() +3* df.RestingBP.std())]
df1.shape
```

```python
df2 = df1[df1.RestingBP>=(df1.RestingBP.mean() -3 * df1.RestingBP.std())]
df2.shape
```

```python
df2[df2.Cholesterol>(df2.Cholesterol.mean() +3* df2.Cholesterol.std())]
```

```python
df2[df2.Cholesterol<(df2.Cholesterol.mean() -3* df2.Cholesterol.std())]
```

### no outliers in negative

```python
df3 = df2[df2.Cholesterol<=(df2.Cholesterol.mean() +3 * df.Cholesterol.std())]
df3.shape
```

```python
df3[df3.MaxHR >(df3.MaxHR.mean() +3 * df3.MaxHR.std())]
```

```python
df3[df3.MaxHR <(df3.MaxHR.mean() -3 * df3.MaxHR.std())]
```

```python
df4 = df3[df3.MaxHR >= (df3.MaxHR.mean() -3 * df3.MaxHR.std())]
df4.shape
```

## df4 is our final

```python
df4.shape
```

```python
df.shape
```

```python
df4.head()
```

```python
df4_w_d = pd.get_dummies(df4,drop_first=True,dtype=int)
```

```python
df4_w_d.astype(int)
```

```python
df4_w_d.head()
```

```python
df4_w_d.shape
```

```python
X = df4_w_d.drop('HeartDisease',axis=1)
```

```python
y= df4_w_d['HeartDisease']
```

```python
X.shape
```

```python
y.shape
```

## so we are done know use 1st Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
Logistic = LogisticRegression(max_iter=2000)
```

```python
X_train , X_test ,y_train , y_test = train_test_split(X,y,test_size=0.2)
```

```python
Logistic.fit(X_train,y_train)
```

```python
Logistic.score(X_test,y_test)
```

# so using the Logistic we get this score 0.8296703296703297
